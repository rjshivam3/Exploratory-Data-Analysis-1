## Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?


```python
wine_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 12))
for i, feature in enumerate(wine_data.columns[:-1], 1):  
    plt.subplot(4, 3, i)
    sns.histplot(wine_data[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew
import numpy as np

skewness = wine_data.apply(lambda x: skew(x.dropna()))
print("Skewness of Features:\n", skewness)

non_normal_features = skewness[skewness.abs() > 1].index
print("Non-Normal Features:", non_normal_features)

transformed_features = wine_data.copy()

for feature in non_normal_features:
    if feature in transformed_features.columns:
       
        transformed_features[feature] = np.log1p(transformed_features[feature])

plt.figure(figsize=(15, 12))
for i, feature in enumerate(non_normal_features, 1):
    plt.subplot(4, 3, i)
    sns.histplot(transformed_features[feature], kde=True, bins=30)
    plt.title(f'Transformed Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
features = wine_data.drop(columns=['quality'])
target = wine_data['quality']

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

features_scaled[:5]
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA()
pca.fit(features_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(cumulative_explained_variance, marker='o')
plt.axhline(y=0.90, color='r', linestyle='--')
plt.title('Cumulative Explained Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

num_components_90 = (cumulative_explained_variance >= 0.90).argmax() + 1
num_components_90
```
